```python
import matplotlib.pyplot as plt
```

```python
## Simple Plot
plt.plot([1,2,3],[3,5,1])

plt.show()
```

```python
plt.plot([1,5,9],[3,9,1])
plt.title("Information")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show()
```

```python
## Bar graph is used for comparision for categories

plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,30,70,80])
plt.title("Information")
plt.show()
```

```python
## Histogram

age = [3,2,5,5,10,35,40,32,45,18,18,22,25,50,19,18,17]
bins = [0,5,10,15,20,25,30,35,40,45,50]
plt.hist(age,bins,histtype='step')
plt.show()
```

```python
## Scatter Plot, is generally used to compare the variables, How variables effect.

x = [2,2.5,3,3.5,4,4.5]
y = [8,8.5,9,9.5,10,10.5]

x1 = [8,8.2,8.3,8.6,9]
y1 = [3,3.1,3.4,3.5,4]

plt.scatter(x,y, color='r',label='Line 1')
plt.scatter(x1,y1, color='b',label='Line 2')
plt.legend()
plt.plot()
```

```python
import pandas as pd

iris = pd.read_csv("/content/drive/MyDrive/Salaries/Iris.csv")

iris.head()
```

```python
for species,irissubset in iris.groupby('Species'):
  plt.scatter(irissubset['PetalLengthCm'],irissubset['PetalWidthCm'],label=species)
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for species,irissubset in iris.groupby('Species'):
  ax.scatter(irissubset['PetalLengthCm'],irissubset['PetalWidthCm'], irissubset['SepalLengthCm'],label=species)
plt.legend()
ax.set_xlabel('PetalLength')
ax.set_ylabel('PetalWidth')
ax.set_zlabel('SepalWidth')
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

def f(t):
  return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)

t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(222)  #no of plots, row wise or column wise, 1 2 plot

plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(221)

plt.plot(t2, np.cos(2*np.pi*t2))

plt.show()
```
